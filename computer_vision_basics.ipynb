# Computer vision and Image processing introducion

# how an image is representated

* the range 0 to 255 has to do with how computers store 8-bit numbers 
* we can always divide all the values by 255 to normalize to between 0 and 1
* color images represented Red, Green and Blue combination
* each color channel will have intensity values.
* The shape of the color array then has 3 dimensions 
    * Height, Widht, No.of channels - here 3 : RGB.
    * 720, 1280, 3-RGB

# Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

## read image

```python
pic = Image.open('/kaggle/input/cat-and-dog/training_set/training_set/cats/cat.1002.jpg')
pic
```

```python
# check the type after opening the image using PIL
type(pic)
```

```python
# convert into arrays
pic_arr = np.asarray(pic)
pic_arr
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

## pick a red color only

```python
pic_red = pic_arr.copy()
pic_red
```

```python
plt.imshow(pic_red)
```

```python
# R G B
pic_red[:,:,0] 
```

```python
# default matplotlib applies color schema and below we will apply grey
plt.imshow(pic_red[:,:,0])
```

```python
# RED Channel values 0 is no red and 255 is full red
plt.imshow(pic_red[:,:,0], cmap='gray')
```
